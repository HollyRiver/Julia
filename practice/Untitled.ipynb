```julia
using Plots, PlutoUI, Distributions, Random
```

```julia
N = 10000
U1 = rand(N)
U2 = rand(N)

X = @. sqrt(-2*log(1-U1))*cos(2*pi*U2)
histogram(X); histogram!(randn(N))
```

```julia
N = 10000
R = sqrt.(rand(Exponential(2), N))
theta = rand(N) * 2pi
X = (@. R*sin(theta))
Y = (@. R*cos(theta))

histogram(X); histogram!(Y)
```
